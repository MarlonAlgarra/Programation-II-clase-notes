```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t*3)
s1 = .6 + np.cos(1 * np.pi * t*3)

fig, ax = plt.subplots()
ax.plot(t, s,t,s1)

ax.set(xlabel='Eje X', ylabel='Eje Y',
       title='Ejemplo')
ax.grid()

fig.savefig("test.png")
plt.show()
```

```python
x=np.arange(-2.0,4.0,0.05)
s2=x**3
s3=(s2>3)*4; #print('y4=',s3)
s4=np.zeros(len(x))+5;
fig, ax1 = plt.subplots()
ax1.plot(x, s2,x,s3,x,s4)

ax1.set(xlabel='Eje X', ylabel='Eje Y',
       title='Ejemplo')
ax1.grid()

fig.savefig("test1.png")
plt.show()
```

```python
x=np.arange(-2.0,4.0,0.05)
#s2=x**3
s3=1+np.sin(2*np.pi*x)*np.exp(-x); #print('y4=',s3)
cte=1.1
s4=np.zeros(len(x))+cte;
s5=(s3>cte)*cte*2;
fig, ax1 = plt.subplots()
ax1.plot(x, s3,x,s4,x,s5)

ax1.set(xlabel='Eje X', ylabel='Eje Y',
       title='Ejemplo')
ax1.grid()

fig.savefig("test1.png")
plt.show()
```
