# Tarea: Intersecciones entre gráficas

Primero, las librerías necesarias para realizar la gráfica propuesta:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Generamos entonces los datos para la gráfica y hallamos su promedio:

```python
dt = 0.05
t = np.arange(-2.0, 4.0, dt)
limite = 8
n = 10
r = np.random.randint(-limite,limite+1,n) 
print('Datos aleatorios\n',r)
p1 = np.mean(r)
```

Ahora, indicamos la información para que muestre la gráfica por pantalla, además de generar el 'for' que identifica los cruces entre la linea del promedio y la gráfica de datos aleatorios punto por punto:

```python
fig, ax = plt.subplots()
ax.plot(r,'o-')
ax.set(xlabel='Eje X', ylabel='Eje Y',title='Números aleatorios')
ax.grid(); ax.axis([0,n,-limite-1,limite+1])
ax.plot([0,n],[p1,p1])
ax.text(n/2,p1+.5,'Prom='+str(p1));
for i in range(0,len(r)-1):
    if r[i]<p1<r[i+1] or r[i]>p1>r[i+1]:
        x1=((p1-r[i])/(r[i+1]-r[i])*((i+1)-i))+i
        ax.plot(x1,p1,'o-')
```
